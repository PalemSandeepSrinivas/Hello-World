```python
from openpyxl import Workbook
from openpyxl.styles import PatternFill

def create_colored_table(data, output_file='colored_table.xlsx', fill_color='00FFFF00'):  # Light yellow
    """
    Creates an Excel table with a specified background color for all cells.

    Args:
        data (list): A list of lists, where each inner list represents a row of data.
        output_file (str): The name of the output Excel file.
        fill_color (str): The color to fill the cells with (in hexadecimal format). 
    """
    wb = Workbook()
    ws = wb.active

    for row in data:
        ws.append(row)

    # Apply fill color to all cells
    for row in ws.iter_rows():
        for cell in row:
            cell.fill = PatternFill(start_color=fill_color, end_color=fill_color, fill_type='solid')

    wb.save(output_file)

# Sample data
data = [
    ["Name", "Age", "City"],
    ["Alice", 25, "New York"],
    ["Bob", 30, "London"],
    ["Charlie", 22, "Paris"]
]

create_colored_table(data)
```

```python
from openpyxl import Workbook
from openpyxl.styles import PatternFill

def create_multiple_colored_tables(data_list, output_file='multi_tables.xlsx', 
                                  fill_colors=['00FFFF00', '00FF0000', '0000FFFF']):
    """
    Creates multiple colored tables side-by-side in the same Excel sheet.

    Args:
        data_list (list): A list of lists, where each inner list represents a table 
                         (each table should be a list of lists, where each inner 
                         list is a row of data).
        output_file (str): The name of the output Excel file.
        fill_colors (list): A list of colors (in hexadecimal format) for each table.

    """
    wb = Workbook()
    ws = wb.active
    start_col = 1  # Starting column for the first table

    for i, data in enumerate(data_list):
        fill_color = fill_colors[i % len(fill_colors)]  # Cycle through colors

        for row in data:
            ws.append(row)

        # Apply fill color to the current table
        for row in ws.iter_rows(min_row=ws.min_row, 
                                max_row=ws.max_row, 
                                min_col=start_col, 
                                max_col=start_col + len(row) - 1):
            for cell in row:
                cell.fill = PatternFill(start_color=fill_color, 
                                       end_color=fill_color, 
                                       fill_type='solid')

        start_col += len(row[0]) + 2  # Move to the next table's starting column

    wb.save(output_file)

# Sample data (multiple tables)
data_list = [
    [["Name", "Age", "City"],
     ["Alice", 25, "New York"],
     ["Bob", 30, "London"],
     ["Charlie", 22, "Paris"]],
    [["Product", "Quantity"],
     ["Laptop", 10],
     ["Phone", 20],
     ["Tablet", 5]],
    [["Country", "Population"],
     ["USA", 330],
     ["India", 1400], 
     ["China", 1450]]
]

create_multiple_colored_tables(data_list)
```

```python
def print_table(data):
  """
  Prints a simple table to the console.

  Args:
    data: A list of lists, where each inner list represents a row of data.
           The first inner list should contain the column headers.
  """

  # Get column widths
  col_widths = []
  for col_idx in range(len(data[0])):
    col_widths.append(max(len(str(row[col_idx])) for row in data))

  # Print header
  header = " | ".join(f"{col:>{width}}" for col, width in zip(data[0], col_widths))
  print("-" * (len(header) + 2))
  print(f"| {header} |")
  print("-" * (len(header) + 2))

  # Print data rows
  for row in data[1:]:
    row_str = " | ".join(f"{item:>{width}}" for item, width in zip(row, col_widths))
    print(f"| {row_str} |")

  print("-" * (len(header) + 2))

# Sample data
data = [
    ["Col1", "Col2"],
    [1, 4],
    [2, 5],
    [3, 6]
]

# Print the table
print_table(data)
```

```python
from datetime import date

def calculate_age(birthdate):
  """
  Calculates the age of a person given their birthdate.

  Args:
    birthdate: A string representing the birthdate in the format 'YYYY-MM-DD'.

  Returns:
    The age of the person in years as an integer.
  """
  try:
    born = date.fromisoformat(birthdate)
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
  except ValueError:
    print(f"Invalid birthdate format. Please use 'YYYY-MM-DD' format.")
    return None

# Example usage:
birthdate = "1997-07-23"
age = calculate_age(birthdate)

if age is not None:
  print(f"Age: {age} years")
```

```python
CREATE TABLE RandomPeople (
    Name VARCHAR(50),
    Occupation VARCHAR(50)
);

INSERT INTO RandomPeople (Name, Occupation)
SELECT 
    CASE rn 
        WHEN 1 THEN 'John' 
        WHEN 2 THEN 'Jane' 
        WHEN 3 THEN 'David' 
        WHEN 4 THEN 'Mary' 
        WHEN 5 THEN 'Michael' 
        WHEN 6 THEN 'Sarah' 
        WHEN 7 THEN 'Christopher' 
        WHEN 8 THEN 'Lisa' 
        WHEN 9 THEN 'Daniel' 
        WHEN 10 THEN 'Jennifer' 
        ELSE 'Other' 
    END,
    CASE 
        WHEN rn % 10 = 0 THEN 'Doctor' 
        WHEN rn % 10 = 1 THEN 'Engineer' 
        WHEN rn % 10 = 2 THEN 'Teacher' 
        WHEN rn % 10 = 3 THEN 'Programmer' 
        WHEN rn % 10 = 4 THEN 'Artist' 
        WHEN rn % 10 = 5 THEN 'Writer' 
        WHEN rn % 10 = 6 THEN 'Scientist' 
        WHEN rn % 10 = 7 THEN 'Businessman' 
        WHEN rn % 10 = 8 THEN 'Chef' 
        WHEN rn % 10 = 9 THEN 'Musician' 
        ELSE 'Other' 
    END
FROM (
    SELECT 
        ROW_NUMBER() OVER (ORDER BY (SELECT NULL)) AS rn 
    FROM 
        sys.objects 
    WHERE 
        ROW_NUMBER() OVER (ORDER BY object_id) <= 10 
) AS RowNumbers;
```
